```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("../Feynn Labs/crop_reco.csv")
```

```python
df
```

```python
df.isnull().sum()
```

```python
sns.relplot(x='ph',y='humidity',data=df,hue='label',size='rainfall',aspect=1.5,height=8)
pt.savefig('rain.png')
```

```python
pt.figure(figsize=(20,12),dpi=200)
sns.pairplot(data=df,hue='label',vars=['humidity','temperature','ph','rainfall'])
pt.savefig('variables.png')
```

```python
pt.figure(dpi=200)
sns.pairplot(data=df,hue='label',vars=['N','P','K'],aspect=1.5)
pt.savefig('NPK.png')
```
